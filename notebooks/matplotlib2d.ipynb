```python
import matplotlib.colors as mcolors
def catcm(data, cmap1, cmap2, centre=0, N=100):
    # https://stackoverflow.com/questions/30082174/join-two-colormaps-in-imshow
    cmap = {name:plt.get_cmap(name) for name in (cmap1, cmap2)}
    vmin, vmax = (np.nextafter(min(data.min(), -1), -np.inf), 
                  np.nextafter(max(data.max(), 1), np.inf))
    levels = np.concatenate([np.linspace(vmin, centre, N, endpoint=False),
                             np.linspace(centre, vmax, N+1, endpoint=True)])
    colors = np.concatenate([cmap[name](np.linspace(0, 1, N)) 
                             for name in (cmap1, cmap2)])
    cmap, norm = mcolors.from_levels_and_colors(levels, colors)
    return cmap, norm
cmap, norm = catcm(A, 'cmo.ice', 'cmo.turbid_r')
plt.imshow(data, cmap=cmap, norm=None)
bar = plt.colorbar()
```

## imshow

### Basics

Images are $M\times N$ arrays [1]. The `plt.imshow(A)` function shows images like arrays i.e. so that the first index (representing rows) increases along the vertical axis (top row, ..., bottom row),  and the second index (representing columns) increases along the horizontal axis (left column, ..., right column).

In other words: `image[x,y]=A[y,x]`. This gives a *map* (in a cartographic sense) of `A.T`, not `A` as desired. To get a map of `A` one should call `imshow(A.T)` because `image[x,y]=A.T[y,x]=A[x,y]`.

In contrast, vectors / points / coordinates are plotted with the first index increasing along the horizontal axis and the second index along the vertical axis. 

Thus, `A` has to be transposed.

---
[1] They may also be represented by $M\times N \times 3$ or $M\times N \times 4$ arrays  (2d arrays storing 3 RGB values and, in the latter case, 1 transparency value).

### Adjust the colorbar

```python
def cbar(im):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    divider = make_axes_locatable(plt.gca())
    cax = divider.append_axes("right", size="3%", pad=0.2)
    cbar = plt.colorbar(im, cax=cax)   
im = plt.imshow(a)
cbar(im)
```

### Clip the colormap 

```python
im = plt.imshow(a, vmin=0.5, vmax=0.6)
cbar(im)
```

### Rescale the axes

```python
im = plt.imshow(a, extent=[-100,100, 4e3,5e3])
cbar(im)
```

This tries to set the aspect-ratio to equal for both axes, and I chose random ranges => that's why it looks squashed. In real life it'll rarely be the case. But we still can work around this:

```python
im = plt.imshow(a, extent=[-100,100, 4e3,5e3])
plt.gca().set_aspect(0.1)
```

### Apply shading

```python
from matplotlib.colors import LightSource
ls = LightSource(azdeg=45, altdeg=45)
cmap = plt.cm.get_cmap('viridis')
im = ls.shade(a, cmap=cmap, blend_mode='hsv', vert_exag=50)
plt.imshow(im)
```

### Add a mask

We can also darken some parts of the image we don't want to show (e.g. with poor data-coverage => not credible)

```python
dark_min, dark_max = 0.2, 0.8 # everything between will NOT be darkened.
nx, ny = a.shape
greys = np.full((nx,ny,3), 70)
alphas = np.array(a)
alphas = -((alphas <= dark_min) | (alphas >= dark_max)).astype(int) * 0.5 + 1
im = plt.imshow(alphas)
cbar(im)
```

Note that we converted booleans (0 or 1) to integers to perform arithmetics on it

```python
cmap = plt.cm.get_cmap('viridis')
a_rgb = cmap(a) # convert our array to RGB with an additional channel for transparency
print(a_rgb.shape, '=> the last dimension is R,G,B,alpha')
a_rgb[..., -1] = alphas # put our alphas into that 4th channel
# the above is a handy shortcut for:
# a_rgb[:, :, 3] = alphas # negative indices mean counting from the last 
plt.imshow(greys) # this is our grey background
plt.imshow(a_rgb) # the bits with alpha < 1 will show the grey background lurking at the back
```

### More on rescaling and `extent`

```python
# data 
row0 = [0,1]
row1 = [2,3]
a = np.array([row0, row1])
np.matrix(a)
```

```python
# default extent  
"""
It plots the matrix A:
matrix([[0, 1],
        [2, 3]])
so that A11 is in the top left corner
of the plot. It seems correct, however
the 

"""
import matplotlib.cm as cm
cmap = cm.get_cmap('hot', len(a.flatten()))
plt.imshow(a, cmap=cmap)

plt.annotate('a[0,0]', (0,0), color='w')
plt.annotate('a[1,0]', (0,1), color='k')
plt.annotate('a[0,1]', (1,0), color='k')
plt.annotate('a[1,1]', (1,1), color='k')
plt.gca().set_xticks([0,1])
plt.gca().set_yticks([0,1])
plt.colorbar()
plt.gca().invert_yaxis()
plt.show()
```

The Y axis has a wrong extent now! `A[0,0]` is no longer in the origin of the plot. Note also that pixels are no longer centred at integers.

```python
# Custom extent 
x1, x2, y1, y2 = 0, 1, 0, 1
extent = np.array([x1, x2, y1, y2])
cmap = cm.get_cmap('hot', len(a.flatten()))
plt.imshow(a, cmap=cmap, extent=extent)
plt.colorbar()
plt.gca().invert_yaxis()
plt.show()
```

## contour/f

### contourf

OK, let's see the contour plots in action too!

```python
# contourf fills the space between discretely spaced contours 
cntr = plt.contourf(a)
plt.gca().invert_yaxis() # we didn't have to do it with imshow!
```

```python
# help(plt.contourf)
```

```python
cntr = plt.contourf(a, levels=2)
plt.gca().invert_yaxis() # we didn't have to do it with imshow!
cbar(cntr)
```

This shows that **imshow is a limit case of contourf for a very large number of contours!**

```python
cntr = plt.contourf(a, levels=100)
plt.gca().invert_yaxis() # we didn't have to do it with imshow!
cbar(cntr)
```

Lastly, the problem we've been struggling with

```python
cntr = plt.contourf(a, levels=5, vmax=0.6)
plt.gca().invert_yaxis() # we didn't have to do it with imshow!
cbar(cntr)
```

One way to work around this is to clip the array before passing to the plotting function

```python
a_clipped = np.clip(a,a_min=np.min(a),a_max=0.6)
```

```python
cntr = plt.contourf(a_clipped, levels=5)
plt.gca().invert_yaxis() # we didn't have to do it with imshow!
cbar(cntr)
```
